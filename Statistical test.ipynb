I saw this meme on Twitter and wanted to know if the samples were statistically different: turns out they were!
image.png

```python



import numpy as np
import scipy.stats as stats

# Review A
review_A = np.clip(np.random.normal(5, 0, 19), a_min=0, a_max=5)

# Review B
review_B = np.clip(np.random.normal(4.6, 5, 2280), a_min=0, a_max=5)

# MW test
stat, p_value = stats.mannwhitneyu(review_A, review_B)
print(stat, p_value)

```
